## Preprocessing

#### Encoding Categorical Variable (Species):

```python
encoder = LabelEncoder()
df['Species'] = encoder.fit_transform(df['Species'])
```

#### Splitting Data into Features and Target:

```python
X = df.drop(['Id', 'Species'], axis=1)
Y = df['Species']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

#### Scaling Numeric Variables:

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

## Model Training and Evaluation

#### Model Training - Support Vector Machine (SVM):

```python
svm = SVC(kernel='linear', C=1.0, random_state=42)
svm.fit(X_train, y_train)
```

#### Model Evaluation:

```python
predictions = svm.predict(X_test)
print("Accuracy:", accuracy_score(y_test, predictions))
print("\nClassification Report:\n", classification_report(y_test, predictions))
```

### Saving and Using the Model

#### Saving the Model using Pickle:

```python
with open('svm_model.pkl', 'wb') as f:
    pickle.dump((encoder, scaler, svm), f)
```

#### Predicting with the Saved Model:

```python
final = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])

with open('svm_model.pkl', 'rb') as f:
    encoder, scaler, model = pickle.load(f)

final_scaled = scaler.transform(final)
final_scaled_df = pd.DataFrame(final_scaled, columns=X.columns)

predictions_new = model.predict(final_scaled_df)
predicted_species = encoder.inverse_transform(predictions_new)

print("\nPrediction of Species:")
for i, species in enumerate(predicted_species):
    print(f"Sample {i+1}: {species}")
```
